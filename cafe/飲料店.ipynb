# MilkShop

```python
url = "https://www.milksha.com/store_detail.php?uID=1"
def into_Taipei():
    driver.get(url)
    time.sleep(3)
    # 跳到地區選擇區塊
    target_area = driver.find_elements(By.CSS_SELECTOR, "select[id='area']")
    feed = target_area[0]
    move_element_be_visible(driver, feed)

    # 點選查詢北部店家

    #按下"請選擇縣市"
    button_area = driver.find_elements(By.CSS_SELECTOR, "select[id='area']")[0]
    button_area.click()
    #選擇北部
    button_norst = driver.find_elements(By.CSS_SELECTOR,"select[id='area'] > option[value='14']")[0]
    button_norst.click()
    #按下Go
    button_Go = driver.find_elements(By.CSS_SELECTOR,"button[type='submit']")[0]
    button_Go.click()
    time.sleep(3)
into_Taipei()
```
