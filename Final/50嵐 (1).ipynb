```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, StaleElementReferenceException, ElementNotInteractableException
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import time
import random

def initialize_browser():
    options = webdriver.EdgeOptions()
    options.add_argument("--start-maximized")
    return webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)

def get_busy_times(day_blocks):
    busy_times = []
    for block in day_blocks:
        times = block.find_elements(By.CSS_SELECTOR, "div.dpoVLd")
        for time_elem in times:
            busy_times.append(time_elem.get_attribute("aria-label"))
    return busy_times

def search_store(driver, search_textbox, store_name):
    # 清除搜索欄位
    try:
        clear_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "div.lSDxNd button[aria-label='關閉']"))
        )
        clear_button.click()
        time.sleep(random.uniform(1, 2))
    except (NoSuchElementException, TimeoutException, ElementNotInteractableException, StaleElementReferenceException):
        pass

    search_textbox.send_keys("可不可熟成紅茶-" + store_name)
    search_button = driver.find_element(By.ID, "searchbox-searchbutton")
    search_button.click()

def get_store_name(driver):
    try:
        store_name_element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "h1.DUwDvf.lfPIob"))
        )
        return store_name_element.text
    except (NoSuchElementException, TimeoutException, StaleElementReferenceException):
        return None

def get_business_hours(driver):
    try:
        time_element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//div[contains(@aria-label, '隱藏本週營業時間')]"))
        )
        aria_label = time_element.get_attribute("aria-label")
        start_index = aria_label.find("星期三")
        end_index = aria_label.find("星期日")
        wed_sat_hours = aria_label[start_index:end_index]
        wed_sat_hours = wed_sat_hours.replace("星期四、10:00 到 21:30;", "").replace("星期五、10:00 到 21:30;", "")
        return wed_sat_hours
    except (NoSuchElementException, TimeoutException, StaleElementReferenceException):
        return None

def get_day_busy_times(driver, day_name):
    try:
        day_blocks = WebDriverWait(driver, 20).until(
            EC.visibility_of_all_elements_located((By.CSS_SELECTOR, "div.g2BVhd.eoFzo"))
        )
        busy_times = get_busy_times(day_blocks)
        return busy_times if busy_times else ["無資料"]
    except (NoSuchElementException, TimeoutException, StaleElementReferenceException):
        return ["無資料"]

def click_next_day_button(driver, times):
    for _ in range(times):
        try:
            next_day_button = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "button.jOGDgf.Hk4XGb"))
            )
            next_day_button.click()
            # 添加隱式等待，等待網頁元素加載完成
            driver.implicitly_wait(10)  # 等待時間可以根據需要調整
            time.sleep(random.uniform(1, 3))
        except (NoSuchElementException, TimeoutException, ElementNotInteractableException, StaleElementReferenceException):
            # 再次嘗試獲取並點擊按鈕
            try:
                next_day_button = WebDriverWait(driver, 20).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, "button.jOGDgf.Hk4XGb"))
                )
                next_day_button.click()
                # 添加隱式等待，等待網頁元素加載完成
                driver.implicitly_wait(10)  # 等待時間可以根據需要調整
                time.sleep(random.uniform(1, 3))
            except (NoSuchElementException, TimeoutException, ElementNotInteractableException, StaleElementReferenceException):
                continue
def main():
    driver = initialize_browser()
    driver.get("https://www.google.com/maps")
    time.sleep(10)

    data_list = []

    name_info = [
        '開封店',  '昆陽店', '景美店', '中山南西店', '大橋延三店',
        '北投中和店', '古亭店', '科技大樓店',  '龍山寺店', '中山遼寧店',
        '延吉店',  '內湖江南店', '大安店', '大安信義店',
        '大直店', '天母德行東店', '長春店', '松山饒河店', '忠孝新生店', '石牌裕民店',
        '中正南昌店', '內湖瑞光店', '市政店', '公館店', '錦州店', '南京三民店',
        '信義永春店', '伊通店', '通化店', '士林大北店', '內湖民權店', '晴光店', '台北站前店'
    ]

    search_textbox = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input#searchboxinput"))
    )

    for index, name in enumerate(name_info, start=1):
        print(f"\n========== 店家 {index} ==========")
        search_store(driver, search_textbox, name)
        time.sleep(random.uniform(5, 10))

        store_name = get_store_name(driver)
        if not store_name:
            print(f"店名： {name}")
            print("找不到店家資訊")
            continue

        print(f"店名： {store_name}")

        business_hours = get_business_hours(driver)
        if business_hours:
            print("星期三和星期六的營業時間（刪除星期四和星期五）：", business_hours)
        else:
            print("無法提取營業時間")
            continue

        print("星期三的時段資訊：")
        wednesday_busy_times = get_day_busy_times(driver, "星期三")
        for time_label in wednesday_busy_times:
            print(time_label)

        click_next_day_button(driver, 3)

        print("星期六的時段資訊：")
        saturday_busy_times = get_day_busy_times(driver, "星期六")
        for time_label in saturday_busy_times:
            print(time_label)

        # 收集每個小時的繁忙程度資料
        hour_data = []
        for hour in range(6, 24):
            busy_level = ""
            if wednesday_busy_times and len(wednesday_busy_times) > hour - 6:
                busy_level += wednesday_busy_times[hour - 6]
            busy_level += "/"
            if saturday_busy_times and len(saturday_busy_times) > hour - 6:
                busy_level += saturday_busy_times[hour - 6]
            hour_data.append(busy_level)

        # 添加到主資料列表中
        data_list.append([index, store_name, *business_hours.split(" "), *hour_data])

    driver.quit()

    # 列印資料列表
    print("\n========== 爬取結果 ==========")
    for data in data_list:
        print(data)

if __name__ == "__main__":
    main()
```
